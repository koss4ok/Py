Импорт библиотек и настройка графиков

```python
import pandas as pd
import seaborn as sbn
import pandas_profiling as pdpf
sbn.set(rc = {'figure.figsize':(15,8)})
```

Чтение данных и удаление "мусорной колонки"

```python
df = pd.read_csv("C:\\Users\\k_makolov\\Desktop\\ds_salaries.csv")
df = df.drop('Unnamed: 0',axis = 1)
df.columns
```

Построение линейного графика по опыту:<br>
<small>EN - начальный уровень (стажёр/джун)<br>
MI - средний уровень (мидл)<br>
SE - высокий уровень (сеньёр)<br>
EX - высший уровень (синьёр плюс/ лид)</small>

```python
grouping_by_expirience = df.groupby('experience_level')['salary_in_usd'].mean()
grouping_by_expirience = grouping_by_expirience.sort_values()
print(grouping_by_expirience)
sbn.lineplot(data = grouping_by_expirience)
```

Данные по размеру компании

```python
grouping_by_size = df.groupby('company_size')['salary_in_usd'].mean()
grouping_by_size = grouping_by_size.sort_values()
print(grouping_by_size)
sbn.barplot(x = grouping_by_size.index, y = grouping_by_size)
```

Данные по странам

```python
grouping_by_country = df.groupby('company_location')['salary_in_usd'].median()
grouping_by_country = grouping_by_country.sort_values()
print(grouping_by_country)
sbn.barplot(x = grouping_by_country.index, y = grouping_by_country)
```

```python
grouping_by_country_emp = df.groupby('employee_residence')['salary_in_usd'].median()
grouping_by_country_emp = grouping_by_country_emp.sort_values()
print(grouping_by_country_emp)
sbn.set(rc = {'figure.figsize':(20,8)})
sbn.barplot(x = grouping_by_country_emp.index, y = grouping_by_country_emp)
```
